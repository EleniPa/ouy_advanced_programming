# Exercise One

```python
# so that a dataframe can be used
import pandas as pd
```

```python
file = open("PeopleTrainngDate.csv")
```

```python
# split into seperate strings in a list by comma, remove unnecessary characters
entry_list = []
for line in file:
    items = line.split(",")
    stripped_items = []
    for i in items:
        stripped_items.append(i.replace('"', "").replace("\n", "").replace(".", "").lstrip(" "))
    entry_list.append(stripped_items)
```

```python
# check for uniformity of length
length_list = []
for l in entry_list:
    x = len(l)
    length_list.append(x)
```

```python
# get column titles as a variable
cols = entry_list.pop(0)
```

```python
#remove last item in entry list - anomalous item
entry_list.pop()
```

```python
# use a for loop to create lists for each column
title = []
first_name = []
surname = []
i_d = []
email = []
company = []
updated = []
for l in entry_list:
    title.append(l[0])
    first_name.append(l[1])
    surname.append(l[2])
    i_d.append(l[3])
    email.append(l[4])
    company.append(l[5])
    updated.append(l[6])
        
```

```python
# create one name variable from surname and first_name using list comprehension
name = [x + ", " + y for x,y in zip(surname, first_name)]
```

```python
#create a dictionary of column title and column variables(containing all values)
title_column_dict = {cols[0]: title, cols[1]: name, cols[2]: i_d, cols[3]: email, cols[4]: company, cols[5]: updated}
```

```python
result_df_ex1 = pd.DataFrame(title_column_dict)
```

```python
result_df_ex1
```

```python
#create dataframe
result_df_ex2 = pd.DataFrame(title_column_dict_rearr)
```

```python
for l in entry_list:
    email_id = l[1].split(" ")
    l.remove(l[1])
    l.extend(email_id)
    
    
```

```python
# use a for loop to create lists for each column
title = []
first_name = []
surname = []
i_d = []
email = []
company = []
updated = []
for l in entry_list:
    title.append(l[1])
    first_name.append(l[3])
    surname.append(l[2])
    i_d.append(l[6])
    email.append(l[5])
    company.append(l[4])
    updated.append(l[0])
```

```python
# list comprehension to create one name variable from surname and first_name
name = [x + ", " + y for x,y in zip(surname, first_name)]
```

```python
# dict following same order as result_df_ex2, which this incoming data will be merged with
title_column_dict_rearr = {cols[5]: updated, cols[0]: title, cols[1]: name, cols[2]: i_d, cols[3]: email, cols[4]: company}
```

```python
# create_dataframe
df_ex3 = pd.DataFrame(title_column_dict_rearr)
```

```python
# merge with existing df
dfs_to_merge = [result_df_ex2, df_ex3]
result_df_ex3 = pd.concat(dfs_to_merge)
```

```python
# convert updated to datetime for sorting
result_df_ex3['Updated'] = pd.to_datetime(result_df_ex3['Updated'])
```

```python
# voila
result_df_ex3.sort_values('Updated')
```
